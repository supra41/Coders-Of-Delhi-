```python
def load_data(filename="data.json"):
    from pathlib import Path
    import json
    cwd = Path.cwd()
    path = Path(filename)
    if not path.is_absolute():
        path = cwd / path
    print(f"Loading JSON from: {path}")  # helps debug where it's looking
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except FileNotFoundError:
        raise FileNotFoundError(f"File not found: {path} (cwd: {cwd})")
    except json.JSONDecodeError as e:
        raise ValueError(f"Invalid JSON in {path}: {e}")
    return data
```

```python
data = load_data("data.json")
```

```python
#write a function to display users and their connections
def display_users(data):
    print("Users and their connections:")
    for user in data['users']:
        print(f"ID:{user['id']}. {user['name']} is frinds with: {user['friends']} and liked pages are {user['liked_pages']}")
    print("\nPages Informations")
    for page in data['pages']:
        print(f"{page['id']}: {page['name']}")
display_users(data)
```
